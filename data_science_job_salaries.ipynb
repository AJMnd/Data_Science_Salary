```python
# Import pandas and numpy

import pandas as pd
import numpy as np
```

```python
# Load csv

file_path = "wages/Data_Science_Jobs_Salaries.csv"
data_salary_df = pd.read_csv(file_path)
data_salary_df
```

```python
# Check counts of job titles
data_salary_df["job_title"].value_counts()
```

```python
data_salary_df["work_year"].unique()
```

```python
# Remove "e" from work year

data_salary_df["work_year"] = data_salary_df["work_year"].str.extract('(\d+)', expand=False)
data_salary_df
```

```python
data_salary_df2 = data_salary_df.drop(columns=['experience_level', 'employment_type', 'remote_ratio', 'company_size', 'employee_residence'])
data_salary_df2
```

```python
data_salary_df2.to_csv("data_science_job_salaries_cleaned.csv", index=False)
```
